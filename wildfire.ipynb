```python
import numpy as np
import pandas as pd
import os
import cv2
# import labelencoder
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
path = 'images'
folders = []
for folder in os.listdir(path):
    folders.append(folder)
print(folders)
```

```python
# convert the folders list to array
folders = np.array(folders)
folders
```

```python
# Use Label Encoder to convert the folders name to numbers
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
labels = le.fit_transform(folders)
```

```python
labels
```

```python
image_array = []
image_label = []

for i in folders:
    path = 'images/' + i
    for image in os.listdir(path):
        image = cv2.imread(path + '/' + image)
        image = cv2.resize(image, (64,64))
        # image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image_array.append(image)
        image_label.append(le.transform([i])[0])
```

```python
image_array = np.array(image_array)
image_label = np.array(image_label)
```

```python
image_array = image_array.astype("float32")/255.0
```

```python
# import train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(image_array, image_label, test_size = 0.2, random_state = 42)
```

```python
import tensorflow as tf
# import layers, models
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
#import model
from tensorflow.keras.models import Sequential
from tensorflow.keras import models, layers
```

```python
model = models.Sequential()

# Convolutional and max-pooling layers
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Dropout layer
model.add(layers.Dropout(0.5))

# Flatten layer
model.add(layers.Flatten())

# Dense layers
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(64, activation='relu'))

# Output layer
model.add(layers.Dense(len(folders), activation='softmax'))  # num_classes is the number of classes you want to predict

model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=20, validation_data=(x_test, y_test)) 
```

```python
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_pred_classes
```

```python
x_test[0]
```

```python
# single_image = # Assuming x_test[0] is a single image
single_image = np.expand_dims(x_test[0], axis=0)

# Make predictions with the model
predictions = model.predict(single_image)
print(predictions)
np.argmax(predictions[0])
```

```python
# Convert one-hot encoded labels to original labels
y_test_original = le.inverse_transform(y_test)
y_pred_original = le.inverse_transform(y_pred_classes)
```

```python
# import confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
cm = confusion_matrix(y_test_original, y_pred_original)
print("Confusion Matrix:")
print(cm)
```
